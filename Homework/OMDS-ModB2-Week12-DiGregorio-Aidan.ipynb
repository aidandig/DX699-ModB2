# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

Advantages:
  - Two separate subgraphs you could clearly see each datapoint without overlapping
  - One graph allows you to see each graph in context with one another

Disadvantages:
  - Two separate subgraphs could make it difficult to compare the overall trends. At least to me, it is easier to compare overall trends of the lines when they are plotted on one graph.
  - Depending on the amount of data, graphing the lines on the same graph could make it difficult to read. However, only 7 points for each lineplot on this example is completely fine to read.

Due to a mix of categorical (encoded features) and numerical (e.g., age) features, K-Prototypes could be used to discover complex clusters of relationships for cancer mortalities: 

https://towardsdatascience.com/the-k-prototype-as-clustering-algorithm-for-mixed-data-type-categorical-and-numerical-fe7c50538ebb/

```python
cancer_deaths_encoded.columns.tolist()
```

```python
from kmodes.kprototypes import KPrototypes
```

```python
continuous_columns = ['AGE']
categorical_columns = [col for col in cancer_deaths_encoded.columns if col != 'AGE' and col != 'DATA_YEAR']

df_sample = cancer_deaths_encoded.sample(frac=0.2, random_state=42)

scaler = StandardScaler()
cancer_deaths_scaled = df_sample.copy()
cancer_deaths_scaled[continuous_columns] = scaler.fit_transform(df_sample[continuous_columns])

cat_indices = [cancer_deaths_scaled.columns.get_loc(col) for col in continuous_columns]

kproto = KPrototypes(n_clusters=5, init='Cao', verbose=1)
clusters = kproto.fit_predict(cancer_deaths_scaled, categorical=cat_indices)

df_sample['Cluster'] = clusters
```

```python
# See the average Age for each cluster
print(df_sample.groupby('Cluster')['AGE'].mean())

# See which Race is most common in each cluster
print(df_sample.groupby('Cluster')[['RACE_Asian', 'RACE_Black', 'RACE_Hispanic', 'RACE_Multiracial', 'RACE_NHOPI', 'RACE_White']].mean())

# Checking size of each cluster
print(df_sample.groupby('Cluster').size())
```

```python
print(f"Shape of categorical centroids: {kproto.cluster_centroids_[1].shape}")
```

```python
centroids_cat_data = kproto.cluster_centroids_[1]
centroids_cat_data
```

```python
centroids_cont = pd.DataFrame(kproto.cluster_centroids_[0], columns=continuous_columns)

# Get the raw data's shape
cat_data_raw = kproto.cluster_centroids_[1]

print(f"Detected Shape: {n_rows} rows, {n_cols} columns")
print(f"Names provided: {len(categorical_columns)}")

# Re-align the data
try:
    # Most likely: 5 rows (clusters), columns matching your list
    n_cat_cols = len(categorical_columns)
    # We take only the first (5 clusters * n_features) values and reshape
    data_reshaped = cat_data_raw[:5 * n_cat_cols].reshape(5, n_cat_cols)
    centroids_cat = pd.DataFrame(data_reshaped, columns=categorical_columns)
except ValueError:
    # If the above fails, let's just make a generic DF to see what the data looks like
    print("Warning: Shape mismatch persists. Slicing data to match column count.")
    centroids_cat = pd.DataFrame(cat_data_raw[:len(categorical_columns)].reshape(1, -1), columns=categorical_columns)

# Final Merge
cluster_profiles = pd.concat([centroids_cont, centroids_cat], axis=1)
print("--- Successfully Aligned Risk Profiles ---")
print(cluster_profiles)
```
