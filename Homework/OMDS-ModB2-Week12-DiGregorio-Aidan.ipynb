# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

Advantages:
  - Two separate subgraphs you could clearly see each datapoint without overlapping
  - One graph allows you to see each graph in context with one another

Disadvantages:
  - Two separate subgraphs could make it difficult to compare the overall trends. At least to me, it is easier to compare overall trends of the lines when they are plotted on one graph.
  - Depending on the amount of data, graphing the lines on the same graph could make it difficult to read. However, only 7 points for each lineplot on this example is completely fine to read.

```python
# filling NaN values as 7 since 1-5: natural, accident, suicide, homicide, pending investigation, 7: could not be determined
cancer_deaths_encoded['MANNER_OF_DEATH_CODED'] = cancer_deaths_encoded['MANNER_OF_DEATH_CODED'].fillna(7)

nans_cancer_deaths = cancer_deaths_encoded[cancer_deaths_encoded.isna().any(axis=1)]
nans_cancer_deaths
```

```python
# converting Boolean features into 1s and 0s
bool_cols = cancer_deaths_encoded.select_dtypes(include=['bool']).columns
cancer_deaths_encoded[bool_cols] = cancer_deaths_encoded[bool_cols].astype(int)
```

```python
nans_cancer_deaths = cancer_deaths_encoded[cancer_deaths_encoded.isna().any(axis=1)]
nans_cancer_deaths
```

```python
# need to handle missing AGE values. More complex than a simple fillna because we don't want to skew the AGE trends

# filling NaN AGEs as median age by SEX for a better imputation than just the total median AGE for entire dataset
cancer_deaths_encoded['AGE'] = cancer_deaths_encoded['AGE'].fillna(cancer_deaths_encoded.groupby('SEX_Male')['AGE'].transform('median'))
print(f"Missing values in AGE: {cancer_deaths_encoded['AGE'].isnull().sum()}")
```

```python
# Verify that there are no remaining samples with missing values

nans_cancer_deaths = cancer_deaths_encoded[cancer_deaths_encoded.isna().any(axis=1)]
nans_cancer_deaths
```

```python
# Drop redundant feature since it is exactly the same as SEX_Male
cancer_deaths_encoded.drop(columns=['SEX_CODED_M'], inplace=True)
```
